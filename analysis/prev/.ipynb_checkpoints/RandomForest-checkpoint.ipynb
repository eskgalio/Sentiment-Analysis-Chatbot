```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, roc_curve, auc
from collections import Counter
```

```python
# Load processed dataset
df = pd.read_csv("Processed_SentimentData.csv")
```

```python
# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(df['cleaned_statement'], df['encoded_status'], test_size=0.2, random_state=42)
```

```python
# TF-IDF Vectorization (Optimized)
vectorizer = TfidfVectorizer(max_features=3000)  # Reduced feature size
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
# Optimized Random Forest Model
rf_model = RandomForestClassifier(n_estimators=50, max_depth=20, random_state=42, n_jobs=-1)  # Optimized params
rf_model.fit(X_train_tfidf, y_train)  # Train the model
```

```python
# Make Predictions
y_pred = rf_model.predict(X_test_tfidf)
accuracy = accuracy_score(y_test, y_pred)
print(f"Optimized Random Forest Accuracy: {accuracy:.4f}")
```

```python
# Save Model and Vectorizer
joblib.dump(rf_model, "random_forest_model.pkl")
joblib.dump(vectorizer, "tfidf_vectorizer.pkl")

print("Optimized Random Forest model and TF-IDF vectorizer saved successfully.")
```

```python
# Load the saved Random Forest model and TF-IDF vectorizer
rf_model = joblib.load("random_forest_model.pkl")
vectorizer = joblib.load("tfidf_vectorizer.pkl")

# Prepare data again (same process as before)
X_test_tfidf = vectorizer.transform(df['cleaned_statement'])
y_test = df['encoded_status']

# Make predictions
y_pred = rf_model.predict(X_test_tfidf)
```

```python
# 1️⃣ **Feature Importance Visualization**
feature_names = vectorizer.get_feature_names_out()
importances = rf_model.feature_importances_

top_n = 15
indices = np.argsort(importances)[-top_n:]

plt.figure(figsize=(10, 5))
sns.barplot(x=[feature_names[i] for i in indices], y=importances[indices], palette="coolwarm")
plt.xticks(rotation=45)
plt.xlabel("Words")
plt.ylabel("Feature Importance")
plt.title("Top 15 Words Influencing Random Forest Model")
plt.show()
```

```python
# 2️⃣ **Effect of Number of Trees on Accuracy**
tree_counts = [5, 10, 20, 50, 100]
accuracies = []

for n in tree_counts:
    rf = RandomForestClassifier(n_estimators=n, max_depth=10, random_state=42, oob_score=True)
    rf.fit(X_test_tfidf, y_test)
    acc = accuracy_score(y_test, rf.predict(X_test_tfidf))
    accuracies.append(acc)

plt.figure(figsize=(8, 5))
plt.plot(tree_counts, accuracies, marker="o", linestyle="-", color="red")
plt.xlabel("Number of Trees")
plt.ylabel("Accuracy")
plt.title("Effect of Number of Trees on Accuracy in Random Forest")
plt.show()
```

```python
# 3️⃣ **Sentiment Category vs. Text Length**
df['text_length'] = df['cleaned_statement'].apply(lambda x: len(x.split()))

plt.figure(figsize=(8, 5))
sns.boxplot(x=df['encoded_status'], y=df['text_length'], palette="coolwarm")
plt.title("Text Length vs. Sentiment Category")
plt.xlabel("Sentiment Category")
plt.ylabel("Number of Words")
plt.show()
```

```python
# 4️⃣ **ROC Curve Analysis**
y_prob = rf_model.predict_proba(X_test_tfidf)[:, 1]  # Get probability scores
fpr, tpr, _ = roc_curve(y_test, y_prob, pos_label=rf_model.classes_[1])
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6, 5))
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Random guess line
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Random Forest")
plt.legend()
plt.show()
```

```python
# 5️⃣ **Comparison of True vs. Predicted Labels**
plt.figure(figsize=(8, 5))
sns.scatterplot(x=y_test, y=y_pred, alpha=0.6, color="green")
plt.xlabel("Actual Label")
plt.ylabel("Predicted Label")
plt.title("Comparison of True vs. Predicted Labels")
plt.show()
```

```python
# 6️⃣ **Sentiment Distribution (Before & After Prediction)**
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Original sentiment distribution
sns.countplot(x=y_test, palette="viridis", ax=axes[0])
axes[0].set_title("Original Sentiment Distribution")
axes[0].set_xlabel("Sentiment Class")
axes[0].set_ylabel("Count")

# Predicted sentiment distribution
sns.countplot(x=y_pred, palette="coolwarm", ax=axes[1])
axes[1].set_title("Predicted Sentiment Distribution")
axes[1].set_xlabel("Sentiment Class")
axes[1].set_ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
# 7️⃣ **Prediction Function**
def predict_sentiment(text):
    text_tfidf = vectorizer.transform([text])  # Convert text to TF-IDF
    prediction = rf_model.predict(text_tfidf)[0]  # Get predicted class
    return prediction

# Example Predictions
sample_texts = [
    "I feel so lost and hopeless.",
    "I'm extremely happy with my progress!",
    "Things are getting worse every day...",
    "Feeling okay, but still uncertain about the future."
]

for text in sample_texts:
    predicted_class = predict_sentiment(text)
    print(f"Text: {text}\nPredicted Sentiment: {predicted_class}\n")
```

```python
# 8️⃣ **Random Forest Prediction Confidence Analysis**
probabilities = rf_model.predict_proba(X_test_tfidf)

plt.figure(figsize=(8, 5))
sns.histplot(np.max(probabilities, axis=1), bins=30, kde=True, color="blue")
plt.title("Random Forest Prediction Confidence Distribution")
plt.xlabel("Highest Probability Score")
plt.ylabel("Number of Samples")
plt.show()
```

```python
# 🔟 **Tree-Level Analysis (First 5 Test Samples)**
tree_predictions = np.array([tree.predict(X_test_tfidf) for tree in rf_model.estimators_])

# Count how many trees predicted each class for first 5 test samples
for i in range(5):
    counts = Counter(tree_predictions[:, i])
    print(f"Test Sample {i+1}: {counts}")
```
